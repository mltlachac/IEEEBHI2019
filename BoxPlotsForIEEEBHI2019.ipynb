```python
#ML Tlachac

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
fileset = ["tRdf.csv", "cRdf2.csv", "cRdf3.csv", "cRdf4.csv", "cRdf0.25.csv", "cRdf0.5.csv", "cRdf0.75.csv", "cRdf0.25w.csv", "cRdf0.5w.csv", "cRdf0.75w.csv", "nRdf.csv"]
nameset = ['1', "C(2)", "C(3)", "C(4)", "C(25%)", "C(50%)", "C(75%)", "C(0.25)", "C(0.5)", "C(0.75)", "All"]
```

```python
dfk = pd.read_csv("resultskNN128.csv")
parameterset = set(dfk.parameter)
for parameter in parameterset:
    print("kNN with k=" + str(parameter))
    df2 = dfk[(dfk.parameter == parameter)]
    df = pd.DataFrame()
    averages = []
    for f in range(0, len(fileset)):
        df[nameset[f]] = list(df2[df2["file"] == fileset[f]].f1List)
        averages.append(sum(df2[df2["file"] == fileset[f]].f1List)/100)
    print(max(averages))
    plt.figure(figsize = (16, 4))
    plt.xlabel("Number of Contacts")
    plt.ylabel("F1")
    plt.title("kNN with k=" + str(parameter))
    df.boxplot(column = ['1', "C(2)", "C(3)", "C(4)", "C(25%)", "C(50%)", "C(75%)", "C(0.75)", "C(0.5)", "C(0.25)", "All"])
    plt.savefig("pics/kNN" + str(parameter) + ".eps")
    plt.show()
    plt.close()
```
